```python
# Since there so many mutations(string values) with unique id or numbers if it has mutation it replaced with 1 or else 0
df1 = df1.applymap(lambda x: '1' if isinstance(x, str) else x)
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np


# Step 1: Standardize your data
scaler = StandardScaler()
df1_scaled = scaler.fit_transform(df1)

# Step 2: Create a dummy target variable (all zeros since it won't be used)
n_samples, _ = df1_scaled.shape
dummy_target = np.zeros(n_samples)

# Step 3: Perform feature selection using SelectKBest
# Specify the number of top features (k) you want to select
k = 200# Adjust the value of 'k' as needed

# Select the top 'k' features using the f_classif scoring function 
selector = SelectKBest(score_func=f_classif, k=k)
scaled_data= selector.fit_transform(df1_scaled, dummy_target)

# Step 4: Apply PCA to the selected features

pca = PCA(n_components=0.95)  # Retain 95% of the variance
reduced_data = pca.fit_transform(scaled_data)

# Now, df1_reduced contains the reduced feature set after both feature selection and PCA.
```

PCA ANALYSIS

```python
reduced_data 
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Create a range of k values to test
k_values = range(2,7)  # You can adjust the range as needed

# Initialize a list to store the sum of squared distances (inertia) for each k
inertia_values = []

# Fit K-means for each k and calculate inertia
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)  # You can change the random_state
    kmeans.fit(reduced_data)
    inertia_values.append(kmeans.inertia_)

# Plot the elbow graph
plt.figure(figsize=(8, 6))
plt.plot(k_values, inertia_values, marker='o', linestyle='-', color='b')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Assuming 'X' is your data
silhouette_scores = []

# Define a range of possible cluster numbers (e.g., from 2 to 10)
for n_clusters in range(2, 11):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(reduced_data)
    
    # Calculate the silhouette score for this clustering
    silhouette_avg = silhouette_score(reduced_data , cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Plot the silhouette scores for different numbers of clusters
plt.figure(figsize=(8, 6))
plt.plot(range(2, 11), silhouette_scores, marker='o', linestyle='--')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score vs. Number of Clusters')
plt.grid(True)
plt.show()
```

```python
# Based on the elbow plot, choose the optimal k (number of clusters)
optimal_k =2# Replace with the value you identified from the elbow plot

# Fit K-means with the optimal number of clusters
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans.fit(reduced_data)

# Get cluster labels for your data points
cluster_labels = kmeans.labels_
```

```python
cluster_labels
```

```python
from scipy.stats import f_oneway
import numpy as np

# Assuming 'reduced_data' is a NumPy array
for feature_index in range(reduced_data.shape[1]):
    feature_values = [reduced_data[cluster_labels == i, feature_index] for i in range(optimal_k)]
    f_statistic, p_value = f_oneway(*feature_values)
    print(f"Feature Index: {feature_index}, F-Statistic: {f_statistic}, p-value: {p_value}")
```

```python
from scipy.stats import f_oneway
import numpy as np

# Create a list to store feature statistics (F-statistic, p-value)
feature_statistics = []

# Assuming 'reduced_data' is a NumPy array
for feature_index in range(reduced_data.shape[1]):
    feature_values = [reduced_data[cluster_labels == i, feature_index] for i in range(optimal_k)]
    f_statistic, p_value = f_oneway(*feature_values)
    
    # Append the feature statistics to the list
    feature_statistics.append((feature_index, f_statistic, p_value))

# Sort the features based on p-values in ascending order
feature_statistics.sort(key=lambda x: x[2])

# Print the features in ascending order of significance (p-value)
for index, f_statistic, p_value in feature_statistics:
    print(f"Feature Index: {index}, F-Statistic: {f_statistic}, p-value: {p_value}")

# Visualize feature importance
# Define the number of top features to plot (change as needed)
num_features_to_plot = 10  # Change the number of top features you want to visualize

# Extract top feature indices and corresponding p-values
top_features = feature_statistics[:num_features_to_plot]
top_feature_indices = [f[0] for f in top_features]
top_feature_p_values = [f[2] for f in top_features]

# Create a bar plot to visualize the top features and their p-values
plt.figure(figsize=(10, 6))
plt.barh(range(num_features_to_plot), [top_feature_p_values[i] for i in range(num_features_to_plot)], align='center')
plt.yticks(range(num_features_to_plot), [top_feature_indices[i] for i in range(num_features_to_plot)])
plt.xlabel('ANOVA p-value')
plt.ylabel('Feature Index')
plt.title('Top Features by ANOVA p-value (Top {} Features)'.format(num_features_to_plot))
plt.gca().invert_yaxis()
plt.show()


```

```python
# Sort features by F-statistic in descending order
feature_statistics.sort(key=lambda x: x[1], reverse=True)

# Select the top N important features (e.g., top 600)
top_features = feature_statistics[:600]
top_features
```

```python
from scipy.stats import mode

for feature_index, _, _ in top_features:
    print(f"Feature {feature_index} Associations with Clusters:")
    for cluster_id in range(optimal_k):
        feature_values_in_cluster = reduced_data[cluster_labels == cluster_id, feature_index]
        mean_value = np.mean(feature_values_in_cluster)
        median_value = np.median(feature_values_in_cluster)
        
        # Calculate the modes (most frequent values) within the cluster
        modes = mode(feature_values_in_cluster)
        mode_values = modes.mode.tolist()
        
        print(f"Cluster {cluster_id}: Mean={mean_value:.2f}, Median={median_value:.2f}, Modes={mode_values}")
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score

# Calculate silhouette score
silhouette_avg = silhouette_score(reduced_data, cluster_labels)

# Calculate Davies-Bouldin index
davies_bouldin = davies_bouldin_score(reduced_data, cluster_labels)

# Print the evaluation metrics
print(f"Silhouette Score: {silhouette_avg}")
print(f"Davies-Bouldin Index: {davies_bouldin}")
```
